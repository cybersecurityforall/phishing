```python
# 5.4.1: Breakdown of Event Counts by Age Group and 2×2 Manipulation.
df.pivot_table(
    index=['age_bin','MANIPULATION'],
    columns='eventName',
    aggfunc='size',
    fill_value=0
)
```

```python
# 5.4.2 and Table 10: Breakdown of Event Counts by Country and 2×2 Manipulation.
df.pivot_table(
    index=['country','MANIPULATION'],
    columns='eventName',
    aggfunc='size',
    fill_value=0
)
```

```python
# Breakdown of Event Counts by Hourly Time Bins.
df.pivot_table(
    index=['EventTimeBin_OneHour'],
    columns='eventName',
    aggfunc='size',
    fill_value=0
)
```

```python
# Breakdown of Event Counts by Day of the Week.
df.pivot_table(
    index=['EventDay'],
    columns='eventName',
    aggfunc='size',
    fill_value=0
)
```
